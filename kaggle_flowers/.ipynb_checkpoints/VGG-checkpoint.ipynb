```python
base_path = './flowers/'
# daisy=1, dandelion=2, rose=3, sunflower=4, tulip=5
categories = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']
```

```python
fnames = []
for category in categories:
    flower_folder = os.path.join(base_path, category)
    file_names = os.listdir(flower_folder)
    full_path = [os.path.join(flower_folder, file_name) for file_name in file_names]
    fnames.append(full_path)
```

```python
print('length for each category:', [len(f) for f in fnames])
```

```python
images = []
for names in fnames:
    one_category_images = [cv2.imread(name) for name in names if (cv2.imread(name)) is not None]
    images.append(one_category_images)
```

```python
def cvtRGB(img):
    return cv2.cvtColor(img.copy(), cv2.COLOR_BGR2RGB)
```

```python
len(images)
```

```python
plt.figure(figsize=(15,10))
for i, imgs in enumerate(images):
    plt.subplot(2,3,i+1)
    idx = np.random.randint(len(imgs))
    plt.imshow(cvtRGB(imgs[idx]))
    plt.grid('off')
    plt.title(categories[i]+' '+str(idx))
plt.show()
```

```python
img_width, img_height = 256, 256

img = images[4][659]
print(img.shape)
resized_img = resize(img, (img_width, img_height, 3))
resized_img2 = cv2.resize(img,(img_width, img_height), interpolation = cv2.INTER_CUBIC)
print(resized_img.shape)
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.title('original image (BGR-channel)')
plt.grid(False)
plt.imshow(img)
plt.subplot(2,2,2)
plt.title('original image (RGB-channel)')
plt.grid(False)
plt.imshow(cvtRGB(img))
plt.subplot(2,2,3)
plt.title('resized by skimage (BGR-channel)')
plt.grid(False)
plt.imshow((resized_img))
plt.subplot(2,2,4)
plt.title('resized by opencv (RGB-channel)')
plt.grid(False)
plt.imshow(cvtRGB(resized_img2))
plt.show()
```

```python
resized_images = []
for i,imgs in enumerate(images):
    resized_images.append([cv2.resize(img, (img_width, img_height), interpolation = cv2.INTER_CUBIC) for img in imgs])
```

```python
len_images = [len(imgs) for imgs in resized_images]
len_images
```

```python
tmp_imgs = []
for imgs in resized_images:
    tmp_imgs += imgs
imagesRGB = np.array(tmp_imgs)
```

```python
im_categories = np.zeros((np.sum(len_images)), dtype='uint8')
for i in range(5):
    if i is 0:
        im_categories[:len_images[i]] = i
    else:
        im_categories[np.sum(len_images[:i]):np.sum(len_images[:i+1])] = i
print(im_categories.shape)
print(imagesRGB.shape)
```

```python
seed = 100
np.random.seed(seed)
np.random.shuffle(im_categories)
np.random.seed(seed)
np.random.shuffle(imagesRGB)
```

```python
im_OneHot  = tf.keras.utils.to_categorical(im_categories)
im_OneHot.shape
```

```python
print(categories[np.argmax(im_OneHot[1])])
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.title(np.argmax(im_OneHot[1]))
plt.grid(False)
plt.imshow(imagesRGB[1])
```

```python
train_x = imagesRGB[:int(len(imagesRGB)*0.8)]/255
train_y = im_OneHot[:int(len(imagesRGB)*0.8)]
val_x = imagesRGB[int(len(imagesRGB)*0.8)+1:]/255
val_y = im_OneHot[int(len(imagesRGB)*0.8)+1:]
print(train_x.shape)
print(train_y.shape)
print(val_x.shape)
print(val_y.shape)
```

```python
input_shape = (256, 256, 3)

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), input_shape=input_shape, padding='same', activation='relu'),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(256, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same',),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(4096, activation='relu'),
    tf.keras.layers.Dense(4096, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])

model.summary()
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=1e-3)
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, epochs=10, validation_data=(val_x, val_y))
```
